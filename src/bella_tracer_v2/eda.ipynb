```python
import pandas as pd
import json
```

```python
with open("synthetic_traces.jsonl", "r") as f:
    data = [json.loads(element) for element in f.readlines()]
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
traces = {}
for index, row in df.iterrows():
    traces[index] = {
        "service": row["root_service"],
        "trace_ids": [trace["trace_id"] for trace in row["raw_trace_json"]],
    }
```
